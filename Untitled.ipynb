# NBA 2019-2020 Season MVP Prediction

Due to the COVID-19 outbreak, most NBA teams ended up playing only around 65 games, instead of the normal 82, during the 2019-2020 season. Despite the current NBA season being cut short, we still have a lot of data from this season’s games as well as from the past.

Using this data, **who would have been the MVP of the NBA's 2019-2020 season?**

### Our Data
We use the player box score and team box score datasets to obtain NBA 2012-2018 season data.

```python
df = pd.read_csv('data/2012-18_playerBoxScore.csv')
teamDf = pd.read_csv("data/2012-18_teamBoxScore.csv")

df = df[df["seasTyp"] == "Regular"]
df["gmDate"] = pd.to_datetime(df["gmDate"])
teamDf["gmDate"] = pd.to_datetime(teamDf["gmDate"])
dates = ["2012-09-30", "2013-09-30", "2014-09-30", "2015-09-30", "2016-09-30", "2017-09-29", "2018-09-28", "2019-09-28"]
dates = pd.to_datetime(dates)
seasons = ['2012', '2013', '2014', '2015', '2016', '2017', '2018']

df['season'] = pd.cut(df['gmDate'], dates, labels=seasons)
df["season"] = df["season"].astype(int)
teamDf['season'] = pd.cut(teamDf['gmDate'], dates, labels=seasons)
teamDf["season"] = teamDf["season"].astype(int)

cols = ["playDispNm", "playMin", "playPos", "playPTS", "playAST", "playTO",
        "playSTL", "playBLK", "playPF", "playFGA", "playFGM", "playFG%", "play2PA", "play2PM", "play2P%", "play3PA",
        "play3PM", "play3P%", "playFTA", "playFTM", "playFT%", "playORB", "playDRB", "playTRB", "season", "teamAbbr"]
playerStats = df[cols]
playerStats = playerStats.groupby(["playDispNm", "season", "teamAbbr"], as_index = False).mean()
playerStats
```

Additionally, to account for the 2019-2020 season, we scrape data from https://www.basketball-reference.com/ 

```python
year = 2020
url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)
html = urlopen(url)
soup = BeautifulSoup(html)

soup.findAll('tr', limit=2)
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
headers = headers[1:]
headers

rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]

player2020Stats = pd.DataFrame(player_stats, columns = headers)
cols = ['Player', 'Tm','Pos', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'FT', 'FTA', 'FT%',
        'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK','TOV','PF','PTS']
player2020Stats = player2020Stats[cols]
player2020Stats
```

To obtain each player's win ratio per season, we use a dataset from https://www.kaggle.com/boonpalipatana/nba-season-records-from-every-year, which gives us each NBA team's season record for each year.

```python
records = pd.read_csv("data/Team_Records.csv")

records["Team"] = records["Team"].str.replace("*", "")
records = records.dropna(how = "all")
records["season"] = records["Season"].str[:4].astype(int)
records = records[records["season"] > 2010]
records = records[["season", "Team", "W/L%"]]

teamDict = {"ATL":"Atlanta Hawks", "BKN":"Brooklyn Nets", "BOS":"Boston Celtics",
"CHA":"Charlotte Hornets", "CHI":"Chicago Bulls",
"CLE":"Cleveland Cavaliers", "DAL":"Dallas Mavericks", "DEN":"Denver Nuggets", "DET":"Detroit Pistons",
"GS":"Golden State Warriors", "HOU":"Houston Rockets",
"IND":"Indiana Pacers",
"LAC":"Los Angeles Clippers", "LAL":"Los Angeles Lakers", "MEM":"Memphis Grizzlies", "MIA":"Miami Heat",
"MIL":"Milwaukee Bucks",
"MIN":"Minnesota Timberwolves", "NO":"New Orleans Pelicans", "NY":"New York Knicks",
"OKC":"Oklahoma City Thunder", "ORL":"Orlando Magic",
"PHI":"Philadelphia 76ers",
"PHO":"Phoenix Suns",
"POR":"Portland Trail Blazers", "SA":"San Antonio Spurs",
"SAC":"Sacramento Kings",
"TOR":"Toronto Raptors",
"UTA":"Utah Jazz",
"WAS":"Washington Wizards"}
test = {v: k for k, v in teamDict.items()}

records["Team"] = records["Team"].map(test)
records = records.rename(columns = {"Team":"teamAbbr"})
records

playerStats.columns
playerStats = playerStats.merge(records, how = "left", on = ["season", "teamAbbr"])
playerStats
```

Using our 2012-2018 season dataframe, we calculate Player Efficiency Rating (PER) and True Shooting (TS) percentage.

```python
def calc_PER(df):
    df["PER"] = (df["playFGM"] * 85.910 + df["playSTL"] * 53.897 + df["play3PM"] * 51.757 + df["playBLK"] * 39.190
                 + df["playORB"] * 39.190 + df["playAST"] * 34.677 + df["playDRB"] * 14.707 - df["playPF"] * 17.174
                 - (df["playFTA"] - df["playFTM"]) * 20.091 - (df["playFGA"] - df["playFGM"]) * 39.190 
                 - df["playTO"] * 53.897) * (1/df["playMin"])
    return df

def calc_TS(df):
    df["TS"] = (df["playPTS"]/(2 * (df["playFGA"] + (.44 * df["playFTA"])))) * 100
    return df

playerStats = calc_PER(playerStats)
playerStats = calc_TS(playerStats)
playerStats
```

We use a dataset from https://www.kaggle.com/danchyy/nba-mvp-votings-through-history to obtain MVP voting data in past seasons. We then merge this MVP voting dataframe with our 2012-2018 season dataframe to determine what percentage of total MVP votes (the "award share") each player obtained in each season.

```python
mvpVoting = pd.read_csv("data/mvp_votings.csv")
mvpVoting["season"] = mvpVoting["season"].str[:4].astype(int)
mvpVoting = mvpVoting[mvpVoting["season"] > 2011]

mvpVoting = mvpVoting[["player", "season", "votes_first", "award_share"]]
mvpVoting = mvpVoting.rename(columns = {"player": "playDispNm"})
mvpVoting = mvpVoting.merge(playerStats, how = "left", on = ["playDispNm", "season"])
mvpVoting = mvpVoting.fillna(0)
mvpVoting
```

We then run the SelectKBest algorithm to see which features predict "award_share" the best.

```python
originalFeatures = ['playFGA', 'play3PA', 'playFTA', 'PER', 'TS', 'playMin', 'playPTS', 'playTRB',
                    'playAST', 'playSTL', 'playBLK', 'playFG%', 'play3P%', 'playFT%', 'W/L%']
x = mvpVoting[originalFeatures].values
y = mvpVoting[['award_share']].values
y = y.reshape(y.shape[0], )
selector = SelectKBest(mutual_info_regression, k=5)
best_feat = selector.fit_transform(x, y)
sorted_indices = np.argsort(selector.scores_)[::-1]
for index in sorted_indices:
    print(f"{originalFeatures[index]}: {selector.scores_[index]:.4f}")
```

We also perform a data visualization of the correlation between "award_share" and other features. 

```python
data = mvpVoting[originalFeatures + ["award_share"]]
fig = plt.figure(figsize=(20,20))

fig.set_facecolor('white')

sns.heatmap(data.corr(), annot=True, cmap="RdYlGn")
```

For the features of our model, we remove "playFGA", "playFTA", and "play3PA" since attempts shouldn't matter when determining MVP-level contributions toa team. We also remove "playFG%", "playFT%", and "play3P%" for the same reason.

```python
features = ['PER', 'TS', 'playMin', 'playPTS', 'playTRB',
                    'playAST', 'playSTL', 'playBLK', 'W/L%', 'award_share']

mvpData = mvpVoting[features]
x = mvpData.values
y = mvpData[['award_share']].values
y = y.reshape(y.shape[0], )

fig = plt.figure(figsize=(14, 14))

fig.set_facecolor('white')

sns.heatmap(mvpData.corr(), annot=True, cmap="RdYlGn")
```

We then assign MVP ranks per player for each season.

```python
player_ranks = []
playerStats = playerStats[["playDispNm", "season", "teamAbbr", "playMin", "playPTS", "playAST", "playSTL", "playBLK", "playTRB", "TS", "PER", "W/L%"]]
tempDf = None
playerRanks = []
for season in mvpVoting["season"].unique():
    print(season)
    singleSeasonDf = mvpVoting[mvpVoting["season"] == season]
    singleSeasonDf.sort_values(by = "award_share", ascending = False, inplace = True)
    playerMvp = [1 if i == 0 else 0 for i in range(len(singleSeasonDf))]
    singleSeasonDf["rank"] = np.arange(0, len(singleSeasonDf))
    singleSeasonDf["mvp"] = playerMvp
    if tempDf is None:
        tempDf = singleSeasonDf
    else:
        tempDf = tempDf.append(singleSeasonDf)

renameMap = {"per":"PER", "ts_pct":"TS", "player":"playDispNm", "ast_per_g":"playAST", "stl_per_g":"playSTL",
                "pts_per_g":"playPTS", "blk_per_g":"playBLK", "win_pct":"W/L%", "mp_per_g":"playMin",
                "trb_per_g":"playTRB"}      
tempDf = tempDf.rename(columns = renameMap)
mvpVoting[mvpVoting["season"] == 2015]
```

After, we train our model using Logistic Regression and Mean Reciprocal Rank.

```python
def mean_reciprocal_rank(preds, actual):
    total_sum = 0.0
    for i in range(len(actual)):
        index = np.where(preds == actual)[0]
        if len(index) == 0:
            total_sum += 0.0
            continue
        total_sum += 1. / (index[0]+1)
    return total_sum / len(actual)

def process_data(df):
    print("process data")
    seasons = df["season"].unique()
    train_columns = ['TS', 'PER', 'playPTS', 'playTRB', 'playAST',
            'playSTL', 'playBLK', 'playMin', 'W/L%']
    target_columns = ['rank']

    model = LogisticRegression(multi_class = "multinomial", solver = "newton-cg", C = 1, n_jobs = -1, max_iter = 2000, verbose = 3)
    totalAcc = []
    print("model")
    for s in mvpVoting["season"].unique():
        print("Season: ", s)
        train = df[df["season"] != s]
        val = df[df["season"] == s].sort_values(by = "award_share", ascending = False)
        X_train = train[train_columns].values
        Y_train = train[target_columns].values
        Y_train = Y_train.reshape(Y_train.shape[0], )
        
        X_val = val[train_columns].values
        Y_val = val[target_columns].values
        Y_val = Y_val.reshape(Y_val.shape[0], )
        print("scaler")
        scaler = MinMaxScaler()
        X_train = scaler.fit_transform(X_train)
        X_val = scaler.fit_transform(X_val)
        print("fit")
        X_train, Y_train = shuffle(X_train, Y_train)
        print(X_train)
        model.fit(X_train, Y_train)
        print("predict")
        Y_pred = model.predict(X_val)
        
        rankedPlayers = np.argsort(Y_pred)[::]
        print(rankedPlayers)
        print(Y_val)
        
        totalAcc.append(mean_reciprocal_rank(rankedPlayers[:5], Y_val[:5]))
        totalAvg = np.average(totalAcc)
        print("\n-----------\n")
        print(f"ACC top 5 is: {totalAvg:.4f}")
        print("\n-----------\n")
process_data(tempDf)
```

Some data cleaning...

```python
renameMap = {"Player":"playDispNm", "AST":"playAST", "STL":"playSTL", "PTS":"playPTS", "BLK":"playBLK", 
             "MP":"playMin", "TRB":"playTRB", "FG":"playFGM", "3P":"play3PM", "ORB":"playORB", "DRB":"playDRB",
             "PF":"playPF", "FTA":"playFTA", "FT":"playFTM", "FGA":"playFGA", "TOV":"playTO", "Tm":"teamAbbr"}      
player2020Stats = player2020Stats.rename(columns = renameMap)
player2020Stats.drop(columns = "Pos", inplace = True)

for i in player2020Stats.columns:
    if i != "playDispNm" and i != "teamAbbr":
        player2020Stats[i] = pd.to_numeric(player2020Stats[i])

player2020Stats = calc_PER(player2020Stats)
player2020Stats = calc_TS(player2020Stats)
player2020Stats
```

Finally, we test our model on the NBA players of the 2019-2020 season.

```python
train_columns = ['TS', 'PER', 'playPTS', 'playTRB', 'playAST', 'playSTL', 'playBLK', 'playMin', 'W/L%']
target_columns = ['rank']
# print(mvpVoting)
train_x = tempDf[train_columns].to_numpy()
train_y = tempDf[target_columns].to_numpy()

train_x, train_y = shuffle(train_x, train_y)
train_y = train_y.reshape(train_y.shape[0], )

test_x = player2020Stats[train_columns].to_numpy()
test_x = np.nan_to_num(test_x)

scaler = MinMaxScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.fit_transform(test_x)

model = LogisticRegression(multi_class = "multinomial", solver = "newton-cg", C = 1, n_jobs = -1, max_iter = 2000, verbose = 3)
model.fit(train_x, train_y)

preds = model.predict_proba(test_x)
sorted_indices = np.argsort([float(item) for item in preds[:,0]])[::-1]
for i, index in enumerate(sorted_indices[:10]):
    print(f"{i+1}. is index: {player2020Stats.iloc[index].playDispNm}, {preds[index,0]}")
```
